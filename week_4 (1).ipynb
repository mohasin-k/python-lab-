```python
#tuples
print("\n===TUPLE EXAMPLE===")
power=(20.5,34,-33.4,23.44)
print("Tuple: ",power)
print("First power: ", power[-1])
#tuple unpacking
a,b,c,d = power
print(f"Unpacked: a={a}, b={b}, c={c}, d={d}")
```
